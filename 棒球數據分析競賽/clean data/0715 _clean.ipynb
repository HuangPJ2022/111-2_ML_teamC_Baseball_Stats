## create new df and select feasible columns

```python
copy_raw = raw.copy()
```

```python
copy_raw = copy_raw[copy_raw['events'].notna()]
```

```python
copy_raw = copy_raw.reset_index(drop=True)
```

```python
copy_raw
```

```python
copy_raw['home'] = 0

for idx in range(0, len(copy_raw)):
    if copy_raw['inning_topbot'][idx] == 'Top':
        copy_raw['home'][idx] = 0
    else:
        copy_raw['home'][idx] = 1
```

```python
on_base = []

for idx in range(0, len(copy_raw['on_1b'])):
    onBase = 0
    
    if np.isnan(copy_raw['on_1b'][idx]) == False:
        onBase += 1
    if np.isnan(copy_raw['on_2b'][idx]) == False:
        onBase += 1
    if np.isnan(copy_raw['on_3b'][idx]) == False:
        onBase += 1

    on_base.append(onBase)
```

```python
score_diffs = []
post_score_diffs = []

for idx, home in enumerate(copy_raw['home']):
    home_score = copy_raw['home_score'][idx]
    away_score = copy_raw['away_score'][idx]
    
    p_home = copy_raw['post_home_score'][idx]
    p_away = copy_raw['post_away_score'][idx]

    
    if home == 1:
        diffs = home_score - away_score
        p_diffs = p_home - p_away
    else:
        diffs = away_score - home_score
        p_diffs = p_away - p_home

    score_diffs.append(diffs)
    post_score_diffs.append(p_diffs)
```

### using "delta_run_expentancy", the expentancy for scoring after the bat, to determine whether the PH succeeded.

```python
outcome = []

for idx in range(0, len(copy_raw)):
    if copy_raw['delta_run_exp'][idx] > 0:
        outcome.append(1)
    else:
        outcome.append(0)

    print(copy_raw['delta_run_exp'][idx])
```

```python
outcome
```

```python
# create new df
df = pd.DataFrame()

df['home'] = copy_raw['home']
df['score_diffs'] = score_diffs
df['post_score_diffs'] = post_score_diffs
df['out_when_up'] = copy_raw['outs_when_up']
df['on_base'] = on_base
df['strikes'] = copy_raw['strikes']
df['balls'] = copy_raw['balls']
df['outcome'] = outcome
```

```python
df
```

```python
exceptions = []

for idx in range(0 ,len(copy_raw)):
    if copy_raw['woba_value'][idx] != 0:
        if df['outcome'][idx] != 1:
            exceptions.append(idx)
            print(idx, copy_raw['events'][idx], copy_raw['woba_value'][idx])
```

```python
count_sac_bunt = 0
for sac_bunt in copy_raw['events']:
    if sac_bunt == 'sac_bunt':
        count_sac_bunt += 1

print(count_sac_bunt)

## 有些犧牲觸擊被系統認為會下降得分機率
```

```python
for ecpt in exceptions:
    df['outcome'][ecpt] = 1

## 全部歸類成代打成功，畢竟有達成進壘的效果，雖然系統認為得分機會下降
```

```python
zero = 0
one = 0
for out in df['outcome']:
    if out == 0:
        zero += 1
    else:
        one += 1

print(zero, one)
## 不算 unbalance data
```
