```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'}

url = 'https://github.com/HuangPJ2022/111-2_ML_teamC_Baseball_Stats/tree/main/Stats/performance/standard%20data'

links = []
names = []

response = requests.get(url, headers=headers)

response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.content, 'html.parser')
soup.encode('utf-8')

urls = soup.find_all(class_ = 'css-truncate css-truncate-target d-block width-fit')

## get raw csv link & players name
for url in urls:
    relative_link = url.find('a')['href']
    relative_link = relative_link.replace('/blob','')
    relative_link = 'https://raw.githubusercontent.com' + relative_link
    links.append(relative_link)

    relative_link = relative_link.replace('https://raw.githubusercontent.com/HuangPJ2022/111-2_ML_teamC_Baseball_Stats/main/Stats/performance/standard%20data/','')
    relative_link = relative_link.replace('(Standard)','')
    relative_link = relative_link.replace('.csv','')
    relative_link = relative_link.replace('%20',' ')
    names.append(relative_link)
```

```python
## print(len(links)) -> 888

df = pd.DataFrame()
for idx, url in enumerate(links):
    tempDf = pd.read_csv(url)

    rep_name = []
    player = names[idx]
    for i in range(0, len(tempDf)):
        rep_name.append(player)

    tempDf.insert(0, "Name", rep_name, True)

    df = pd.concat([df, tempDf], ignore_index=True, sort=False)
```
