## 調整實質薪資至 2023 年的價格

```python
new_salary['former season'] = new_salary['contract start year'].apply(lambda x: x - 1) ## 準備後續合併資料所使用

new_salary.head(20)
```

```python
cpi_1980_li = [82.4, 90.9, 96.5, 99.6, 103.9, 107.6, 109.6, 113.6, 118.3, 124.0, 130.7, 136.2, 140.3, 144.5, 148.2, \
          152.4, 156.9, 160.5, 163.0, 166.6, 172.2, 177.1, 179.9, 184.0, 188.9, 195.3, 201.6, 207.3, 215.3, 214.5, 218.1, \
          224.9, 229.6, 233.0, 236.7, 237.0, 240.0, 245.1, 251.1, 255.7, 258.8, 271.0, 292.7, 309.6]

def cpi_given_year(year):
    if int(year) > 2023:
        return cpi_1980_li[2023 - 1980]
    else:
        return cpi_1980_li[year - 1980]

def adjust_cpi(year):
    cpi_year2 = cpi_given_year(2023)
    cpi_year1 = cpi_given_year(year)
    adjusted = 1 * (cpi_year2 / cpi_year1) ## Year 2 Price = Year 1 Price x (Year 2 CPI/Year 1 CPI)
    return adjusted
```

```python
new_salary = new_salary.reset_index(drop=True)
new_salary.head(20)
```

```python
adjusted_wage = []

for idx, year in enumerate(new_salary['contract start year']):
    adjusted_wage.append(new_salary['wage'][idx] * adjust_cpi(year))

adjusted_wage
```

```python
new_salary['adjusted wage'] = adjusted_wage
new_salary['adjusted wage'] = new_salary['adjusted wage'].apply(lambda x: '%d' % x)
```

## 一年多球隊(standard)

#### 做法：把當年度的總成績留下，刪除當年在個別球隊的成績，當年待過的球隊都寫進總成績的Tm欄位

```python
who_be_multiple_team = []
for idx, team in enumerate(standard['Tm']):
    if len(team) == 7:
        who_be_multiple_team.append(standard['Name'][idx])
```

```python
who_be_multiple_team
```

```python
for who in who_be_multiple_team:
    for idx, tm in enumerate(standard.loc[standard['Name'] == who]['Tm']):
        if len(tm) > 3:
            season_look_for = standard.loc[standard['Name'] == who]['Season'].iloc[idx]
            pre_amend = standard.loc[(standard['Name'] == who) & (standard['Season'] == season_look_for)]

            temp_tm = ''
            for i in range(0, len(pre_amend) - 1):
                temp_tm = temp_tm +(pre_amend.iloc[i]['Tm'])
                if i < len(pre_amend) - 2:
                    temp_tm += ', '

            pre_amend.iloc[len(pre_amend) - 1, pre_amend.columns.get_loc('Tm')] = temp_tm
            standard.loc[(standard['Name'] == who) & (standard['Season'] == season_look_for)] = pre_amend
    
standard.loc[standard.Name == 'Oliver Drake']
```

```python
which_year_multiple_tm = []
for idx, team in enumerate(standard['Tm']):
    if len(team) > 3:
        which_year_multiple_tm.append(standard['Season'][idx])
    
which_year_multiple_tm
```

```python
for i in range(0, len(who_be_multiple_team)):
    who = who_be_multiple_team[i]
    which_year = which_year_multiple_tm[i]

    pre_drop =  standard.loc[(standard['Name'] == who ) & (standard['Season'] == which_year )]

    pre_drop = pre_drop.iloc[[-1]]

    standard.loc[(standard['Name'] == who) & (standard['Season'] == which_year)] = pre_drop

standard.loc[standard.Name == 'Oliver Drake']
```

```python
standard = standard[standard['Name'].notna()]
standard =  standard.reset_index(drop=True)
```
