```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
html_text = requests.get('https://baseballsavant.mlb.com/savant-player/yu-darvish-506433?stats=statcast-r-pitching-mlb')
soup = BeautifulSoup(html_text.text, 'lxml')
```

```python
statcast_piching = soup.find('div', id='statcastPitching')
thead = statcast_piching.find('thead')
statcast_piching
```

```python
#爬取欄位名稱
statcast_piching = soup.find('div', id='statcastPitching')
thead = statcast_piching.find('thead')
tr = thead.find('tr')
th = tr.find_all('th')
columns_name = []
for i in th:
    columns_name.append(i.text)
print(columns_name)
```

```python
#爬取data
tbody = statcast_piching.find('tbody')
tr = tbody.find_all('tr', class_="default-table-row")
statcast_data = []
for i in tr:
    td = i.find_all('td')
    statcast_row = []
    for j in td:
        statcast_row.append(j.text.strip())
    statcast_data.append(statcast_row)
statcast_data = statcast_data[:-3]
```

```python
#創建dataframe
df = pd.DataFrame(statcast_data, columns = columns_name)
df
```

```python
#創建dataframe
df = pd.DataFrame(statcast_data, columns = columns_name)
print(df)
```
